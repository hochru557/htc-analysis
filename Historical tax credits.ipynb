```python
# Now let's perform the groupby by the HRC # and then the max of count of the prior rounds applications 
# column to eliminate previous applications from the same participants


round_53_17_latest_round = round_53_17.sort_values(['Count_prior_rounds_applied'], ascending=False).drop_duplicates(subset=["HRC #"], keep="first")
round_53_17_latest_round



```

```python
# Dropping all non-numeric rows

round_53_17_latest_round = round_53_17_latest_round[pd.to_numeric(round_53_17_latest_round['Total Awards'], errors='coerce').notnull()]
round_53_17_latest_round = round_53_17_latest_round[pd.to_numeric(round_53_17_latest_round['20% QRE'], errors='coerce').notnull()]
round_53_17_latest_round = round_53_17_latest_round[pd.to_numeric(round_53_17_latest_round['Total Project Cost'], errors='coerce').notnull()]
round_53_17_latest_round = round_53_17_latest_round[pd.to_numeric(round_53_17_latest_round['Remaining Credit to Award'], errors='coerce').notnull()]
round_53_17_latest_round
```

```python
# Converting the now numeric columns to float data types

round_53_17_latest_round['Total Awards'] = round_53_17_latest_round['Total Awards'].astype(float)
round_53_17_latest_round['20% QRE'] = round_53_17_latest_round['20% QRE'].astype(float)
round_53_17_latest_round['Remaining Credit to Award'] = round_53_17_latest_round['Remaining Credit to Award'].astype(float)
round_53_17_latest_round

```
